```python
import urllib
import re
url = "https://www.upgrad.com/digital-marketing-and-communication-pgc-mica/"
data = urllib.request.urlopen(url)
html_text = data.read().decode('utf-8')
text = re.sub("<.+?>",'',html_text)
print(text)
```
